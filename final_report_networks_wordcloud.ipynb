# Wordclouds

```python
names = ['Dwayne Johnson', 'Chris Hemsworth', 'Robert Downey Jr.', 'Akshay Kumar', 'Jackie Chan',
'Bradley Cooper', 'Adam Sandler', 'Scarlett Johansson', 'Sofia Vergara', 'Chris Evans']
```

```python
# create stopword list:
stop_words = get_stop_words('english')
stop_words.extend(['get', 'gets', 'one', 'two', 'three', 'must', 'take', 
                   'takes', 'find', 'finds', 'look', 'looks', 'man', 'woman',  
                   'will', 'new', 'world', 'small', 'big', 'become', 'young', 'film', 'read', 
                   'family', 'couple', 'friend', 'friends', 'mother', 'father', 'son', 'daughter', 
                   'sister', 'brother', 'boy', 'home', 'help', 'helps', 'stop', 'stops', 
                   'love', 'life', 'year', 'make', 'makes', 'come', 'comes', 'fall', 'falls', 'tries'])
```

```python
def print_wordcloud(actor_name):
    f = open(actor_name + '.txt')
    text = f.read()
    wordcloud = WordCloud(stopwords=stop_words, max_words=50, background_color="white", colormap='seismic').generate(text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(actor_name, fontsize=30)
    plt.show()
```
